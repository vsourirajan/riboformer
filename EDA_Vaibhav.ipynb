```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.optim as optim
import torch.utils.data as data
from scipy.cluster.hierarchy import linkage, dendrogram, leaves_list
from scipy.spatial.distance import squareform
```

```python
data_path = "./data/data_with_human_TE_cellline_all_plain.csv"
df = pd.read_csv(data_path, delimiter="\t")
df.head()
```

```python
#number of unique cell lines -- number of columns that have "bio_source" in the name

print("Number of columns: ", len(df.columns))
print("Number of rows: ", len(df))

na_rows = df[df.isna().any(axis=1)]
print("Number of rows that have NA: ", len(na_rows))

bio_source_cols = [col for col in df.columns if 'bio_source' in col]
print(f"Number of unique human cell lines: {len(bio_source_cols)}")
```

```python
#fill in the null values
df = df.dropna()
na_rows = df[df.isna().any(axis=1)]
print("Number of rows that have NA: ", len(na_rows))
print("Number of rows: ", len(df))
df.head()
```

```python
bio_source_cols = [col for col in df.columns if 'bio_source' in col]
bio_source_df = df[bio_source_cols]
bio_source_df.columns = bio_source_df.columns.str.replace('bio_source_', '')
```

```python
spearman_corr = bio_source_df.corr(method='spearman')
print("Dimensions of spearman correlation matrix: ", spearman_corr.shape)
```

```python
distance_matrix = 1 - spearman_corr
condensed_dist = squareform(distance_matrix.values)
linkage_matrix = linkage(condensed_dist, method='average')

ordered_indices = leaves_list(linkage_matrix)
ordered_corr = spearman_corr.iloc[ordered_indices, ordered_indices]

mask = np.triu(np.ones_like(ordered_corr, dtype=bool))

plt.figure(figsize=(20, 18))
sns.heatmap(
    ordered_corr, 
    mask=mask, 
    cmap='coolwarm', 
    annot=False, 
    fmt=".2f",
    linewidths=0.5, 
    square=True, 
    cbar_kws={"shrink": 0.5}
)

plt.xticks(rotation=90, fontsize=8)
plt.yticks(fontsize=8)
plt.title("Spearman Correlation (Hierarchically Clustered)", fontsize=16)
plt.tight_layout()
plt.savefig('figures/spearman_correlation_heatmap.png')
plt.show()
```

```python
pearson_corr = bio_source_df.corr(method='pearson')
print("Dimensions of pearson correlation matrix: ", spearman_corr.shape)
```

```python
#plot the peason correlation matrix

distance_matrix = 1 - pearson_corr
condensed_dist = squareform(distance_matrix.values)
linkage_matrix = linkage(condensed_dist, method='average')

# Reorder correlation matrix based on hierarchical clustering
ordered_indices = leaves_list(linkage_matrix)
ordered_corr = pearson_corr.iloc[ordered_indices, ordered_indices]

# Mask upper triangle
mask = np.triu(np.ones_like(ordered_corr, dtype=bool))

# Plot heatmap
plt.figure(figsize=(20, 18))
sns.heatmap(
    ordered_corr, 
    mask=mask, 
    cmap='coolwarm', 
    annot=False, 
    fmt=".2f",
    linewidths=0.5, 
    square=True, 
    cbar_kws={"shrink": 0.5}
)

plt.xticks(rotation=90, fontsize=8)
plt.yticks(fontsize=8)
plt.title("Pearson Correlation (Hierarchically Clustered)", fontsize=16)
plt.tight_layout()
plt.savefig('figures/pearson_correlation_heatmap.png')
plt.show()
```
